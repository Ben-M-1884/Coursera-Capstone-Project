# Obtain Coordinates for Towns

```python
# Install what is required
! pip install geopy
! pip install folium

#Import Libraries Required for Clustering
import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
# Obtain Coordinates of the towns in Derbyshire and insert them into lists called Latitude and Longitude.
Latitude = []
Longitude = []

# Iterate through each row in the towns dataframe to get the corrdinates of each town listed in the 'Towns' column
for i in range(towns.shape[0]):
    address = str(towns['Towns'][i] + ', Derbyshire, ' + 'England')
    geolocator = Nominatim(user_agent='ny_explorer', timeout = 3)
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    Latitude.append(latitude)
    Longitude.append(longitude)
    
# The Coordinates will be added to the original dataframe after being clusterd
```

# Cluster Towns into two groups

```python
# K-means clustering will be used to cluster the neighborhoods into 5 clusters
# set number of clusters
kclusters = 2


towns_grouped_clustering = towns.drop('Towns', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(towns_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels to original Dataframe
towns.insert(0, 'Cluster Labels', kmeans.labels_)
# display the dataframe
towns
```

# Create Map of Derbyshire and towns in their cluster groups to the map

```python
# Use the two lists Latitude and Longitude to Create to create a dictionary and then convert the dictionary into a dataframe
dict = {'Latitude' : Latitude, 'Longitude' : Longitude}
Lat_Long_df = pd.DataFrame(dict)
# Combine the 'Lat_Long_df' dataframe to the 'towns' dataframe
towns['Latitude'] = Lat_Long_df['Latitude']
towns['Longitude'] = Lat_Long_df['Longitude']
towns
```

```python
# Obtain Coordinates of Derbyshire
derbyshire_address = 'Derbyshire, England'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(derbyshire_address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Derbyshire are {}, {}.'.format(latitude, longitude))

# create map of Derbyshire using latitude and longitude values
map_Derbyshire = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to map for Each Prospective Town in Derbyshire
for lat, lng, Towns, cluster in zip(towns['Latitude'], towns['Longitude'], towns['Towns'], towns['Cluster Labels']):
    label = '{}'.format(Towns)
    label = folium.Popup((label ), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7,
        parse_html=False).add_to(map_Derbyshire)  

map_Derbyshire
```
